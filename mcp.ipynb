### Model Context Protocol
#### MCP is an open protocol that standadizes how applications provide context to LLMs. Think of MCP like a USB-C port for AI applications. MCP provides a standardized way to connect AI models to different data sources and tools.




- With LLM also, we can connect to MCP and then connect to various tools/services (tool calling) such as Wikipedia, ArxiV, Tavily search, etc., etc.
